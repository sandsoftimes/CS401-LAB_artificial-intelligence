# Problem: Implement the best-first search algorithm on the map such that for any given city, it

```python
import heapq

# Define the map of Romania with distances between cities
romania_map = {
    'Arad': {'Zerind': 75, 'Timisoara': 118, 'Sibiu': 140},
    'Zerind': {'Oradea': 71, 'Arad': 75},
    'Timisoara': {'Lugoj': 111, 'Arad': 118},
    'Sibiu': {'Oradea': 151, 'Arad': 140, 'Fagaras': 99},
    'Oradea': {'Zerind': 71, 'Sibiu': 151},
    'Lugoj': {'Timisoara': 111, 'Mehadia': 70},
    'Fagaras': {'Sibiu': 99, 'Bucharest': 211},
    'Rimnicu Vilcea': {'Sibiu': 80, 'Pitesti': 97, 'Craiova': 146},
    'Mehadia': {'Lugoj': 70, 'Drobeta': 75},
    'Pitesti': {'Rimnicu Vilcea': 97, 'Craiova': 138, 'Bucharest': 101},
    'Craiova': {'Rimnicu Vilcea': 146, 'Pitesti': 138, 'Drobeta': 120},
    'Drobeta': {'Mehadia': 75, 'Craiova': 120},
    'Giurgiu': {'Bucharest': 90},
    'Neamt': {'Iasi': 87},
    'Iasi': {'Neamt': 87, 'Vaslui': 92},
    'Vaslui': {'Iasi': 92, 'Urziceni': 142},
    'Eforie': {'Hirsova': 86},
    'Hirsova': {'Eforie': 86, 'Urziceni': 98},
    'Urziceni': {'Hirsova': 98, 'Vaslui': 142, 'Bucharest': 85},
    'Bucharest': {'Giurgiu': 90}
}

# Define the heuristic values for each city
heuristic_values = {
    'Arad': 366,
    'Mehadia': 241,
    'Bucharest': 0,
    'Neamt': 234,
    'Craiova': 160,
    'Oradea': 380,
    'Eforie': 161,
    'Pitesti': 100,
    'Fagaras': 176,
    'Rimnicu Vilcea': 193,
    'Dobreta': 242,
    'Timisoara': 329,
    'Hirsova': 151,
    'Urziceni': 80,
    'Iasi': 226,
    'Vaslui': 199,
    'Lugoj': 244,
    'Zerind': 374,
    'Sibiu': 253
}

def best_first_search(start_city, goal_city):
    # Create an empty priority queue
    queue = []
    # Enqueue the start city with priority based on the heuristic value
    heapq.heappush(queue, (heuristic_values[start_city], start_city, []))
    
    # Create an empty visited set
    visited = set()
    
    # Create a dictionary to store the shortest distances
    shortest_distances = {city: float('inf') for city in romania_map}
    shortest_distances[start_city] = 0
    
    while queue:
        # Dequeue the city with the highest priority (based on heuristic value)
        _, current_city, path = heapq.heappop(queue)
        
        # Check if we have reached the goal city
       
                # Check if we have reached the goal city
        if current_city == goal_city:
            # Build the final path by combining the current path and the goal city
            final_path = path + [current_city]
            # Calculate the total distance of the path
            total_distance = shortest_distances[current_city]
            # Return the total distance and the final path
            return total_distance, final_path
        
        # Check if the current city has been visited
        if current_city not in visited:
            # Mark the current city as visited
            visited.add(current_city)
            
            # Explore the neighbors of the current city
            for neighbor, distance in romania_map[current_city].items():
                # Calculate the tentative distance from the start city to the neighbor
                tentative_distance = shortest_distances[current_city] + distance
                
                # Update the shortest distance if the tentative distance is smaller
                if tentative_distance < shortest_distances[neighbor]:
                    shortest_distances[neighbor] = tentative_distance
                    # Calculate the priority of the neighbor based on the heuristic value
                    priority = heuristic_values[neighbor]
                    # Enqueue the neighbor with the priority and the updated path
                    heapq.heappush(queue, (priority, neighbor, path + [current_city]))

    # If the goal city is not reachable from the start city, return None
    return None

# Example usage:
start_city = 'Neamt'
goal_city = 'Bucharest'
result = best_first_search(start_city, goal_city)

if result:
    total_distance, path = result
    print(f"Shortest path from {start_city} to {goal_city} is {total_distance} miles and goes as follows:")
    for i in range(len(path)-1):
        print(f"{path[i]} to {path[i+1]}")
else:
    print(f"There is no path from {start_city} to {goal_city}.")
```
