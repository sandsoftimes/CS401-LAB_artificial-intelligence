```python
import pandas as pd
import numpy as np
```

```python
filename='Cust_Segmentation.csv'
```

```python
df=pd.read_csv(filename)
```

```python
df
```

```python
df.head()
```

```python
x=df
x.drop_duplicates(inplace=True)
x.head()
```

```python
# x.dropna(inplace=True)
x=x.dropna(inplace=False)
x.tail()
```

```python
x['Address']
```

```python
x['Address']=pd.to_numeric(df['Address'],errors='coerce')
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
x_scaled
```

```python
x_scaled=pd.DataFrame(x_scaled,columns=x.columns)
```

```python
x_scaled
```

```python
x_scaled.head()
```

```python
y=x_scaled[['Income','Years Employed']]
```

```python
y.head()
```

```python
from sklearn.cluster import KMeans
```

```python
import matplotlib.pyplot as plt
```

```python
KCS=[]
for k in range(1,10):
    kmeans=KMeans(n_clusters=k)
    kmeans.fit(y)
    KCS.append(kmeans.inertia_)
```

```python
plt.plot(range(1,10),KCS)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('KCS')
plt.show
```

```python
kmeans=KMeans(n_clusters=5,random_state=42)
kmeans.fit(y)

y['cluster']=kmeans.labels_
y
```

```python
plt.scatter(y['Income'],y['Years Employed'],c=y['cluster'],cmap='viridis')
plt.xlabel('Income')
plt.ylabel('Years Employeed')
plt.title('Customer Segmentation')
plt.show()
```
